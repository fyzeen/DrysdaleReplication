```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
features150_train_path = "/Users/fyzeen/FyzeenLocal/GitHub/DrysdaleReplication/code/03_FeatureSelection/02_rCCA/results/CC_StabilityTests/150features_trainCC_stability.csv"
features150_test_path = "/Users/fyzeen/FyzeenLocal/GitHub/DrysdaleReplication/code/03_FeatureSelection/02_rCCA/results/CC_StabilityTests/150features_testCC_stability.csv"
features1839_train_path = "/Users/fyzeen/FyzeenLocal/GitHub/DrysdaleReplication/code/03_FeatureSelection/02_rCCA/results/CC_StabilityTests/1839features_trainCC_stability.csv"
features1839_test_path = "/Users/fyzeen/FyzeenLocal/GitHub/DrysdaleReplication/code/03_FeatureSelection/02_rCCA/results/CC_StabilityTests/1839features_testCC_stability.csv"
```

```python
features150_train = pd.read_csv(features150_train_path)
features150_test = pd.read_csv(features150_test_path)
features1839_train = pd.read_csv(features1839_train_path)
features1839_test = pd.read_csv(features1839_test_path)
```

```python
train = plt.violinplot(features1839_train[["reg_CC1", "unreg_CC1"]], positions=[0.5, 1.5])
test = plt.violinplot(features1839_test[["reg_CC1", "unreg_CC1"]])

plt.xticks([0.75, 1.75], ["Regularized CCA", "Normal CCA"])
plt.title("1839 Features")
plt.legend([train['bodies'][0], test['bodies'][0]], ["Train CC1", "Test CC1"])

plt.axhline(0, color="black")


plt.yticks(np.arange(-0.3, 1.1, 0.1))
plt.show()
```
